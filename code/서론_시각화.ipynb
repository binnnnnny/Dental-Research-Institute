## 2. 연령대 별 인구 수 추이

```python
연령인구 = pd.read_excel('/content/인구.xlsx', sheet_name=1)
연령인구
```

```python
# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(연령인구['연령별'], 연령인구['인  구'], color='skyblue')

plt.title('연령별 인구 분포')
plt.xlabel('연령별')
plt.ylabel('인구 수')

plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)

plt.show()
```

```python
age_cleaned = 연령인구.drop(index=0)
# 연령대별 인구 수
age_groups = dict(zip(age_cleaned['연령별'], age_cleaned['인구']))

# 총 인구 수
total_population = sum(age_groups.values())

# 중위수 인구 수
median_population = total_population / 2

# 누적 인구 수를 계산하여 중위수 계산
cumulative_population = 0
median_age_group = None

for age_group, population in age_groups.items():
    cumulative_population += population
    if cumulative_population >= median_population:
        median_age_group = age_group
        break

print(f"중위수에 해당하는 연령대: {median_age_group}")
```
